```python
numerical_columns = ['postal_code', 'model_year', 'electric_range', 'base_msrp', 'legislative_district', 'dol_vehicle_id', '2020_census_tract']

categorical_columns = ['vin_(1-10)', 'county', 'city', 'state', 'make', 'model', 'electric_vehicle_type', 'cafv_eligibility', 'vehicle_location', 'electric_utility']
```

```python
import plotly.express as px
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for column in numerical_columns:
    plt.figure(figsize=(8,6))
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot for {column}")  # Fixed string concatenation
    plt.grid()
    plt.show()
    print(f"Displayed boxplot for {column}")
```

```python
for column in numerical_columns:
    plt.figure(figsize=(8,6))
    sns.kdeplot(x=df[column])
    plt.title(f"kdeplot for {column}")  # Fixed string concatenation
    plt.grid()
    plt.show()
    print(f"Displayed kdeplot for {column}")
```

```python
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column, order=df[column].value_counts().index)  # Order by counts
    plt.title(f"Count Plot for {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.xticks(rotation=90)  # Rotate x-ticks by 90 degrees
    plt.show()
    print(f"Displayed count plot for {column}")
```

```python
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    counts = df[column].value_counts()
    sns.barplot(x=counts.index, y=counts.values)
    plt.title(f"Bar Plot for {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.xticks(rotation=90)  # Rotate x-ticks if they overlap
    plt.show()
    print(f"Displayed bar plot for {column}")
```

```python
for column in categorical_columns:
    plt.figure(figsize=(8, 6))

    # Get the top five categories by count
    counts = df[column].value_counts().nlargest(5)  # Get top 5 categories
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)

    plt.title(f"Pie Chart for Top 5 Categories in {column}")
    plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
    plt.show()

    print(f"Displayed pie chart for top 5 categories in {column}")
```

```python
import plotly.express as px

# Grouping the data by state and counting the number of electric vehicles
state_data = df.groupby('state')['vin_(1-10)'].count().reset_index()
state_data.columns = ['state', 'ev_count']

# Creating the choropleth map
fig = px.choropleth(
    state_data,
    locations='state',
    locationmode='USA-states',
    color='ev_count',
    color_continuous_scale="YlGn",
    scope='usa',
    labels={'ev_count': 'Number of EV Vehicles'},  # Corrected labels syntax
    title='Number of Electric Vehicles by State'
)

# Show the figure
fig.show()
```

```python
ev_made_year = df.groupby(['model_year', 'make']).size().reset_index(name='ev_count')

# Pivoting the DataFrame for the bar chart race
pivot_df = ev_made_year.pivot(index='model_year', columns='make', values='ev_count').fillna(0)

# Creating the bar chart race
bcr.bar_chart_race(
    df=pivot_df,
    title='Number of Electric Vehicles by Car Make Over Years',
    filename='ev_bar_chart_race.mp4',  # Save the animation as a video file
    steps_per_period=10,  # Number of frames to show per time period
    period_length=1000,  # Length of time to display each period in milliseconds
    sort='desc',  # Sort the bars in descending order
    n_bars=5,  # Number of bars to display
    period_label={'x': 0.99, 'y': 0.9, 'ha': 'right', 'va': 'center'},  # Position of the period label
    bar_label_size=7,  # Size of the labels on the bars
    tick_label_size=7,  # Size of the tick labels
    figsize=(5, 3)  # Figure size
)
```
